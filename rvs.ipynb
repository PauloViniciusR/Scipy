```python
from scipy import stats
import numpy as np
```

### O que faz

rvs significa algo como random variates.

Ele serve para simular observações que seguem aquela distribuição.

```python
stats.norm.rvs(size=5)
```

```python
stats.norm.rvs(loc=70, scale=10, size=5)
```

```python
notas = stats.norm.rvs(loc=70, scale=10, size=1000)
print(notas[:10])
```

```python
print(np.mean(notas))
```

```python
print(np.std(notas))
```

```python
print(np.max(notas))
```

```python
print(np.min(notas))
```

### Simulando número de chamadas com Poisson

```python
chamadas = stats.poisson.rvs(mu=4, size=200)
print(chamadas)
```

```python
rng = np.random.default_rng(42)
amostra = stats.norm.rvs(loc=70, scale=10, size=5, random_state=rng)
print(amostra)
```
