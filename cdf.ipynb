```python
from scipy import stats
import numpy as np
```

### O que faz

cdf significa cumulative distribution function.

Ela responde:

P(X ≤ x)

Ou seja:
“Qual a probabilidade da variável ser menor ou igual a esse valor?”

```python
rv = stats.norm(loc=70, scale=10)
prob = rv.cdf(80)
print(prob)
```
