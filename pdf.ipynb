```python
from scipy import stats
import numpy as np
```

### pdf() — para distribuições contínuas

### O que faz

pdf significa probability density function.

Ela devolve a densidade da distribuição naquele ponto.

```python
stats.norm.pdf(70, loc=70, scale=10)
```

```python
rv = stats.norm(loc=70, scale=10)

print(rv.pdf(70))
print(rv.pdf(90))
```

```python
rv = stats.norm(loc=70, scale=10)

print("Densidade em 60:", rv.pdf(60))
print("Densidade em 70:", rv.pdf(70))
print("Densidade em 80:", rv.pdf(80))
print("Densidade em 100:", rv.pdf(100))
```

Usar pdf() para:

- entender o formato da distribuição
- comparar onde a curva é mais alta
- construir gráficos
- analisar a densidade em certos pontos

### pmf() — para distribuições discretas

### O que faz

pdf significa probability density function.

Ela devolve a densidade da distribuição naquele ponto.

```python
stats.poisson.pmf(3, mu=4)
```

```python
stats.binom.pmf(6, n=10, p=0.5)
```

```python
rv = stats.poisson(mu=2)

print("P(X=0)", rv.pmf(0))
print("P(X=1)", rv.pmf(1))
print("P(X=2)", rv.pmf(2))
print("P(X=5)", rv.pmf(5))
```

Diferença mental entre pdf() e pmf()

### pdf:
- contínua
- densidade
- não é probabilidade exata do ponto

### pmf:
- discreta
- probabilidade exata
- faz sentido perguntar P(X = 3)
